## 3.2 Matplotlib

First, we need to bring up the seaborn library, where the Anscombe dataset can be found. We import seaborn using import, and for simplicity make sure it is brought up as sns. 

```python
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
print(anscombe)
```

Matplotlib is Python's main plotting library. We need to import it to use its various features, which are found in the subfolder pyplot, as specified in the code. We will abbreviate it as plt. We are going to change the color of the resulting plot to red instead of blue. The plot function takes one vector for the values of x and the corresponding y-values vector. Plt.plot will, by default, draw lines. 

```python
import matplotlib.pyplot as plt
```

```python
dataset_1 = anscombe[anscombe['dataset'] == 'I']
plt.plot(dataset_1['x'], dataset_1['y'], 'red')
```

If we want the plot to draw circles rather than lines we can include a parameter 'o.' This way the plt.plot uses points instead of lines. The next plot, thus, uses points rather than lines. Also, we readjusted the scale on the y-axis so that it starts from zero rather than 4. 

```python
plt.plot(dataset_1['x'], dataset_1['y'], 'o', 0)
```

To use the other three datasets, we can create other subsets of the anscombe data. 

```python
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

We could make the plots individually or combine them in a grid to see them all together in a single figure. The subplot function takes three parameters:number of rows in the figure for subplots, number of columns in the figure for subplots, and subplot location. The location is sequentially numbered and plots are placed first from left to right and then from top to bottom. 

The following code, which uses the plt.figure() function, will only present an empty figure since all we have done is create a figure and split the figure into a 2 Ã— 2 grid where plots can be placed; no plots have been inserted yet. 

```python
fig = plt.figure()
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
```

If we want to add a plot to each axes we can use the plot method on each axis. Here, we specify that the x variable of dataset 1 should be plotted in relation to the x axis, while the y varibale of dataset 1 should be plotted in relation to the y axis; we have specified that we want to plot points on the upper two plots and x's on the lower two plots. 

```python
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'x')
axes4.plot(dataset_4['x'], dataset_4['y'], 'x')
fig
```

To add labels, we use the tight_layout to make sure the axes are spread apart from one another. To add a title to each plot and for the entire figure, we add the set_title function and the respective plot name and the suptitle function and the overall figure's title. 

```python
axes1.set_title("Dataset 1")
axes2.set_title("Dataset 2")
axes3.set_title("Dataset 3")
axes4.set_title("Dataset 4")
fig.suptitle("Anscombe Data")
fig.tight_layout()
fig
```

We can observe from this plots that even though they have the same summary statistics, the relationships between the data points vary greatly. We can also add labels to each graph's axis. We must remember that the difference between axes and axis is that  each individual subplot plot has axes. The axes contain both an x-axis and a y-axis. All four subplots together make the figure. In the following example, we named the axis of each plot as either y (vertical) or x (horizontal).

```python
axes1.set_xlabel('X')
axes2.set_xlabel('X')
axes3.set_xlabel('X')
axes4.set_xlabel('X')
axes1.set_ylabel('Y')
axes2.set_ylabel('Y')
axes3.set_ylabel('Y')
axes4.set_ylabel('Y')
fig.tight_layout()
fig
```

## 3.3 Statistical Graphics Using Matplotlib
The next set of vizualizations come from the seaborn library. This data set contains the amount of the tips that people leave for various variables. First, we need to load this data set as we did before. 

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

### 3.3.1 Univariate 
Refers to a single variable. 

#### 3.3.1.1 Histogram 
Histograms are great to examine a single variable. The values are "binned," that is they are grouped together and plotted to show the distribution of the variable. In here, we used the hist() function to create our histogram and changed the number of bins from 10 to 20 and made its color brown. The resulting histogram shows that the total bill median is about 35. 

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(tips['total_bill'], bins=20, color='brown')
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
fig.show()
```

### 3.3.2 Bivariate
When we have two variables. 

#### 3.3.2.1 Scatterplot
Scatterplots are great to visualize continuous variables plotted against other continuous variables. We use the scatter() function. The following scatterplot shows a positive relationship between the Total Bill and the Tip, i.e. as the bill increases so does the tip left. In here, we changed the color of the plot to green. 

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'],c='green', cmap=True)
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show()
```

#### 3.3.2.2 Boxplot
Boxplots plot a continuous variable against a dicrete variable and allow us to compare medians, maximums, minimums and ranges. To plot them in Python, the first argument is the data and since there are multiple pieces of data we put each into a list. We then can add labels to the data. In the following boxplot, we added a notch argument that indicates a confidence interval for the median.

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)
axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
     tips[tips['sex'] == 'Male']['tip']],
    labels=['Female', 'Male'], notch=True)
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
boxplot.show()
```

### 3.3.3 Multivariate Data
There are many ways to visualize multivariate data. For example, if we want a scatterplot to present two variables like differences between male and female tipping habits, we can have the men's dots to be presented in one color and size and the female's dots in another to be able to better see how they differ. First, we create a color variable based on sex, using recode_sex() and specifying that if female return 1 and if male (i.e.) else return 0 --this is the change we are making to this plot. Then we can also add arguments to change the size of the dots based on the size of the party. In this case, we multiply the values by 20 to make them bigger and make the differences more apparent. After, we set the color for the sex using c= and set the alpha to 0.8 so that points are more transparent. This way we will better visualize overlapping points. 

```python
def recode_sex(sex):
    if sex == 'Female':
       return 1
    else:
       return 0
tips['sex_color'] = tips['sex'].apply(recode_sex)
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] * 20,
    c=tips['sex_color'],
    alpha=0.8)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show()
```

## 3.4 Seaborn
The matplotlib is the core foundational plotting tool in Python, while seaborn builds on it to provide g a higher-level interface for statistical graphics, producing more aesthetically pleasing and complex visualizations. First, we load the the previous dataset. 

```python
tips = sns.load_dataset("tips")
```

### 3.4.1 Univariate
#### 3.4.1.1 Histograms
With seaborn, histograms are created using sns.distplot. The first line of code is a shortcut for creating separate figure objects and adding individual subplots to the figure. This way we can plot a histogram with the density function. Then, the second line uses distplot to create the actual plot. We use different colors between the histogram and the density plot to differentiate them. 

```python
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], axlabel='Total Bill', color="green", kde_kws={"color":"blue"})
ax.set_title('Total Bill Histogram with Density Plot')
plt.show() 
```

By default, the distplot function will plot the histogram and the density plot. If we only want the former, we need to set the kdw argument to false. Here, as well, we are looking at the histogram in step type, also changing the color to red. 

```python
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False, hist_kws={"histtype": "step", "linewidth": 3,"alpha": 1, "color": "r"})
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

#### 3.4.1.2 Density Plot
We can also visualize an univariate distribution. They present a normal distribution centered at each data point, and then smooth out the overlapping plots so that the area under the curve is 1. In this case, we are chaging the line color to black, increasing its width and adding a label called density.

```python
den, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False, kde_kws={"color": "k", "lw": 5, "label": "Density"})
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
plt.show()
```

#### 3.4.1.3 Rug Plot
A rug plot is a one-dimensional representation of a variableâ€™s distribution and are usually presented with other plots to enhance visualizaiton. In this case, we made the histogram a lighter tone of green to differentiate it from the rug plot below, and made the density plot blue and thinner.  

```python
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True, rug_kws={"color": "darkgreen"}, kde_kws={"color": "b", "lw":2, "label": "Density"}, hist_kws={"alpha": 0.3, "color": "g"})
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
plt.show()
```

#### 3.4.1.4 Count Plot (Bar Plot)
Bar plots are used to count discrete variables instead of binning values to produce a distribution. For this, we use the countplot function. In this case, we are looking at days Thursday through Sunday and the frequency of tips, specifying the difference between male and female customers, also adding a small key to know what each color represents. 

```python
count, ax = plt.subplots()
ax = sns.countplot('day', data=tips, palette="Set3", hue='sex')
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
plt.show()
```

#### 3.4.2.1 Scatterplot
To create a scatterplot in seaborn we use the regplot() function. This function will also plot the line of best fit (regression line), unless we set the fit_reg argument to False. In this case, we change the dots to little, orange pentagons.

```python
scatter, ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, marker='p', scatter_kws={"color":"orange"})
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
```

We can also use lmplot, which calls the more general regplot to create the scatter plot with the line of best fit. In here, we have changed the height and aspect ratio of the resulting plot. Also, the color of the points so that it is easier to differentiate them from the line. 

```python
fig = sns.lmplot(x='total_bill', y='tip', data=tips, height=7, aspect=1, scatter_kws={"color":"black"})
plt.show()
```

We can use jointplot to create a scatterplot that includes a univariate plot on each axis. We must keep in mind that jointplot does not return axes, so we do not need to create a figure with axes on which to place our plot. Instead, this function creates a JointGrid object. In here, we changed the color of the graph and also the space between the main scatterplot and the marginal axes.

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips, color='green', dropna=True, space=0.5)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip',
                fontsize=10, y=1.03)
```

#### 3.4.2.2 Hexbin Plot
Sometimes, scatterplots can be confusing given the number of points they present and the proximity of those points. Thus, sometimes it is better to use a Hexbin Plot, which is similar to a scatterplot with the difference that its bins together those points depending on where they are concentrated. Hexagons are used since they are the most efficient shape for the job. We can use the dropna argument to remove observations that are missing from x and y. We can also change the height and aspect ratio. 

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex", dropna=True, height=6, ratio=4, color='g')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',
                 fontsize=10, y=1.03)
```

#### 3.4.2.3 2D Density Plot
We can also create a 2D density plot, which also plots two variables. These types of plots are specially helpful when we have a lot of points because it avoids overlapping, dividing the plot area in a multitude of small fragments that represent the number of points in this fragment. We can use the cmap argument to create a color map that allows use to better differentiate the shades in our density plot. Here, BuPu instead of just blue creates a color map with blue and purple for better visualization. 

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'],
                 shade=True, cmap="BuPu") 
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
kde_joint = sns.jointplot(x='total_bill', y='tip',
                          data=tips, kind='kde', cmap="BuPu")
```

#### 3.4.2.4 Bar Plot
Bar plots also show multiple variables. By default, barplot will calculate the mean, but the estimator parameter will pass any function like standard deviation or median. In this case, we changed the parameter from mean to median and the color palette to just be shades of blue (Blues_d) since we are looking at different times of the day. Remember that to get the median function working we need to import it from numpy. 

```python
from numpy import median
bar, ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips, estimator=median, palette="Blues_d")
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
plt.show()
```

#### 3.4.2.5 Boxplot
Boxplots show several statistics: the first and third quartiles, the median, the minimums and maximums and the outliers based on the interquartile range. In here, we once again changed the palette to make it more appealing, reduced the width so that the plots are better place in the graph, increased the fliersize to 6 to increase the size of the markers used to indicate outlier observations, and made thinner the width of the gray lines that frame the plot elements by setting fliersize to True. 

```python
box, ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips, palette="Set3", width=0.5, fliersize=6, linewidth=True)
ax.set_title('Boxplot of total bill by time of day')
ax.set_ylabel('Total Bill')
ax.set_xlabel('Time of day')
plt.show()
```

#### 3.4.2.6 Violin Plot
Violin plots are able to show the same values as a boxplot, but plot the â€œboxesâ€ as a density estimation, which might be better at conveying the information of the boxplot for longer term retention. In here, we changed the color paltte to muted and added "sticks" to better visualize how the data is distributed inside the kernel density estimation. We can also use points instead of sticks. 

```python
violin, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, inner="sticks", split=True, palette="muted")
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
plt.show()
```

#### 3.4.2.7 Pairwise Relationships
With mostly numeric data, we can use the pairplot function to visualize all of the pairwise relationships. This function plots a scatterplot between each pair of variables, and a histogram for the univariate data. Here, we have added a plot_kws argument to help differentiate overlapping points in the scatterplots by making the edge color dark blue and the linewidth 1. This helps make better sense of the smaller scale scatterplots. 

```python
fig = sns.pairplot(tips, plot_kws=dict(s=50, edgecolor="darkblue", linewidth=1))
```

We must be careful when using pairplot since the information can be repetitive, i.e. the top two graphs show the same information as the bottom two graphs. Instead, we can use pairgrid to to manually assign the plots for the top half and bottom half and we can use plt.scatter instead of sns.regplot. We can use the despine argument and set to None so that the top and right spines from the plots appears. Once again, since the graphs are relatively small this might help make more sense of them, changing the color of each type of graph to make them more distinctive. 

```python
pair_grid = sns.PairGrid(tips, despine=None)
pair_grid = pair_grid.map_upper(sns.regplot, color='green')
pair_grid = pair_grid.map_lower(sns.kdeplot, color='brown')
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```

#### 3.4.3.3 Facets
Facets are great to show more variables or if we know which plot we want for our visualization, but we want to make multiple plots over a categorical variable. Facets are designed to satisfy these needs, since instead of you needing to individually subset data and lay out the axes in a figure they do this work for you. To make them work, we need Tidy data. In the following example we added the robust argument to estimate a robust regression and deweight outliers, and increased the height of each plot. To create this visualization we simply passed two more parameters into the scatterplot function in seaborn. The col parameter is the variable that the plot will facet by, and the col_wrap parameter determines how many columns we want. However, since we eliminated this latter parameter, the plots appear in the same row rather than in a 2x2 matrix. 

```python
anscombe_plot = sns.lmplot(x='x', y='y', data=anscombe,
                           fit_reg=False,
                           col='dataset', height=3.5, robust=True)
```

The differences between lmplot and regplot is that the former is a figure-level function, while the latter takes into account that many of the plots we created in seaborn are axes-level functions, i.e. not every plotting function will have col and col_wrap parameters for faceting. Therefore, we must first generate a FacetGrid that knows which variable to facet on, and then supply the individual plot code for each facet. The first line of code creates the FacetGrid using that function, while the second line specifies that for each value in time it should plot a histogram of total bill. The following example, increases the height of the resulting graphs and also changes the color of the rugs to red to make them more distinctive. 

```python
facet = sns.FacetGrid(tips, col='time', height=6)
facet.map(sns.distplot, 'total_bill', rug=True,rug_kws={"color": "red"})
plt.show()
```

The following example uses FacetGrid in data samples that are univariate. In the following example, we changed the color palette and also presented the graphs in a 2x2 matrix rather than in a single row, and increased the aspect ration to 2.  

```python
facet = sns.FacetGrid(tips, col='day', hue='sex', palette="Set1", col_wrap=2, aspect=2)
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()
plt.show()
```

The same plot can be created with the lmplot function. This plot adds the line of best fit and also presents the plots in 2 columns at an increased height; otherwise they are too small to make sense of. 

```python
fig = sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=True,
                 hue='sex', col='day', col_wrap=2, height=6)
plt.show()
```

With facets, we can have one variable be faceted on the x-axis, and another variable faceted on the y-axis. We accomplish this by passing a row parameter. This plot adds a legend to know what the blue and red dots mean and also increases the height to make it more understandable.

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex', height=4, palette="Set1")
facet.map(plt.scatter, 'total_bill', 'tip').add_legend()
plt.show()
```

If all of the hue elements overlap, you can employ violin plots so that they do not show this behavior. For this we use the sns.catplot function. The following facet shows violin plots, changing the height and the aspect ratio to make them more easily digestible and also adding margin titles, which shows the titles for the row variables.

```python
facet = sns.catplot(x='day', y='total_bill', hue='sex', data=tips, height=4, aspect=1.8,
                       row='smoker', col='time', kind='violin', margin_titles=True, style="ticks")
```

### 3.5 Panda Objects 
The plotting functions built into Pandas are just wrappers around matplotlib with preset values.

#### 3.5.1 Histograms
Histograms can be created using the Series.plot.hist (first one) or DataFrame.plot.hist function (second one).The first one creates a histogram on a series, while the second one creates a plot on a dataframe, setting an alpha channel transparency so that we can see over the overlapping bars. In the first one, we changed the color and made it a step filled graph rather than a bar graph. We also added a label for the x axis and a title for the graph. For the second one, we added a title and increased the alpha to make the colors more intense. 

```python
fig, ax = plt.subplots()
ax = tips['total_bill'].plot.hist(color="darkgreen", histtype ="stepfilled")
ax.set_title('Total Bill Frequency')
ax.set_xlabel('Total Bill')
plt.show()

fig, ax = plt.subplots()
ax = tips[['total_bill', 'tip']].plot.hist(alpha=1, bins=20, ax=ax)
ax.set_title('Total Bill and Tip Frequency')
plt.show()
```

#### 3.5.2 Density Plot
We created the kernel density estimation plot with the DataFrame.plot.kde function. Here we have changed the color of the graph and also specified the scalar bandwith; we must take into account that using a small bandwidth value can lead to over-fitting, while using a large bandwidth value may result in under-fitting. We also added a tittle and a name to the x-axis. 

```python
fig, ax = plt.subplots()
ax = tips['tip'].plot.kde(color="red", bw_method=0.3)
ax.set_title('Tip Density Estimation')
ax.set_xlabel('Tip')
plt.show()
```

#### 3.5.3 Scatterplot
Scatterplots are created using the DataFrame.plot.scatter function, as the example below shows. Here we also added labels to to the horizontal and vertical axes and a title plus made the color of the points dark blue. 

```python
fig, ax = plt.subplots()
ax = tips.plot.scatter(x='total_bill', y='tip', ax=ax, c="DarkBlue")
ax.set_title('Relation Between Total Bill and Tip')
ax.set_ylabel('Total Bill')
ax.set_xlabel('Tip')
plt.show()
```

#### 3.5.4. Hexbin Plot 
Hexbin plots are created using the Dataframe.plt.hexbin function. We have added a title and proper labels to the x and y axes and also marginals, i.e. we have plotted the marginal density as colormapped rectangles along the bottom of the x-axis and left of the y-axis.

```python
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', ax=ax, marginals=True)
ax.set_title('Relation Between Total Bill and Tip')
ax.set_ylabel('Total Bill')
ax.set_xlabel('Tip')
plt.show()
```

Clearly the example above is hard to make sense of due to its size. Thus, we can use the gridsize parameter to adjust the grid size and better visualize our plot. Here we set it to 13. We also changed the edge colors of the pixels to no color so that we can better see where the data points are clustered, and added a title and labels. 

```python
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', gridsize=12, ax=ax, edgecolors=None)
ax.set_title('Relation Between Total Bill and Tip')
ax.set_ylabel('Total Bill')
ax.set_xlabel('Tip')
plt.show()
```

#### 3.5.5 Boxplot
Boxplots are created with the DataFrame.plot.box function. In this example, we made the boxplots horizontally and organized them in a different order, with the total bill first, then the tip and thne the size. We also changed the width to 0.7 to make it easier to understand and added a title.

### 3.6 SEABORN THEMES AND STYLES
We can change the plot style with the sns.set_style function; this function is run once at the top of your code and the all subsequent plots will use the same style set. The styles that come with seaborn are darkgrid, whitegrid, dark, white, and ticks. The first plot, for comparison, uses no particular style, but the second plot is using a darkgrid style. We have also added a title to both and changed their palettes. 

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill',
                    hue='sex', data=tips,
                    split=True, palette="Set1")
ax.set_title('Violin Plots for Total Bill Comparison Between Male and Females During Dinner and Lunch')
plt.show()
sns.set_style('darkgrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill',
                    hue='sex', data=tips,
                    split=True, palette="Set2")
ax.set_title('Violin Plots for Total Bill Comparison Between Male and Females During Dinner and Lunch')
plt.show()
```

The following code shows what all the styles look like. As we can see, the first two contain a grid, while the second two do not; the first and the third are dark and the second , fourth and fifth are white. Here we decreased the saturation of the plots and included inside the violines the quartiles, and added a hue, palette, and split arguments. Ideally the plots would be larger so that the legend does not interfere in the visualizaiton.  

```python
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1 
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill',
                                data=tips, ax=ax, saturation=0.7, inner="quartile", hue="sex", palette="Set2", split=True, scale_hue=True)
fig.tight_layout()
plt.show()
```
